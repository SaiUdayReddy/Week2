```python
"""
A small sample for Apriori Algorithm in Python
"""
# Creating  Sample Transaction
transactions = [
    ['Milk', 'Bread', 'Saffron'],
    ['Milk', 'Safron'],
    ['Bread', 'Saffron', 'Wafer'],
    ['Bread', 'Wafer'],
]
transactions
```

```python
# Generating association rules
Rules = list(apriori(transactions, min_support=0.2, min_confidence=0.5))
```

```python
# Extract the created rules
associationRules = utils.extract(Rules)
rules_df = pd.DataFrame(associationRules,columns=['LHS','RHS','Support','Confidence','Lift'])
rules_df
```
